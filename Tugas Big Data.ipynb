### Part 2

```python
data_penduduk = pd.read_csv("Data wilayah.csv")
data_area     = pd.read_excel("data_area.xlsx")
```

```python
data_penduduk
```

```python
data_area
```

```python
data_penduduk['wilayah'] = data_penduduk['wilayah'].str.upper()
```

```python
data_penduduk
```

```python
data_total = data_area.merge(data_penduduk,
              how='inner',
              left_on='Neighborhood',
              right_on='wilayah')
data_total
```

```python
data_total = data_total.drop("Neighborhood", axis=1)
data_total.head()
```

```python
column_change = ['wilayah','Areas Region(km squared)', 'Borough', 'latitude', 'longitude','laki-laki', 'perempuan', 'jumlah']
data_total = data_total.reindex(columns=column_change)
```

```python
data_total.head()
```

```python
data_total.columns = ['wilayah','Areas Region(km squared)', 'Borough', 'Latitude', 'Longitude','pria Populasi', 'wanita Populasi', 'Total Populasi']
```

```python
data_total.head()
```

```python
data_total.to_csv("Bali_Full_of_Data.csv", index=False)
```

```python
address = 'Bali, ID'

geolocator = Nominatim(user_agent="id_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Bali are {}, {}.'.format(latitude, longitude))
```

```python
location
```

```python
#Folium Part
map_bli = folium.Map(location=[latitude, longitude], zoom_start=12)
```

```python
map_bli
```

```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
```

```python
# add markers to map
map_bli = folium.Map(location=[latitude, longitude], zoom_start=12, tiles='cartodbpositron')
for lat, lng, borough, neighborhood in zip(data_all['Latitude'], data_all['Longitude'], data_all['Borough'], data_all['wilayah']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='green',
        
        fill_color='#3186cc',
        parse_html=False).add_to(map_bli)  
    
map_bli
```
